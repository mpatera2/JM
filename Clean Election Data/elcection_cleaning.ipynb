```python
import pandas as pd
df = pd.read_csv('electionresults.csv', header = None)
df
```

```python

year_col_map = {
    2024: (2, 3, 4),   # (dem_col, rep_col, ps_label_col)
    2020: (6, 7, 8),
    2016: (10, 11, 13),
    2012: (15, 16, 17),
    2008: (19, 20, 21),
    2004: (23, 24, 25),
    2000: (27, 28, 30),
    1996: (32, 33, 35),
    1992: (37, 38, 40),
    1988: (42, 43, 44),
    1984: (46, 47, 48),
    1980: (50, 51, 53),
    1976: (55, 56, 57),
}

ps_col_map = {
    2024: 408, 2020: 409, 2016: 410, 2012: 411, 2008: 412,
    2004: 413, 2000: 414, 1996: 415, 1992: 416, 1988: 417,
    1984: 418, 1980: 419, 1976: 420
}

national_winners = {
    2024: 'Trump', 2020: 'Biden', 2016: 'Trump', 2012: 'Obama',
    2008: 'Obama', 2004: 'Bush', 2000: 'Bush', 1996: 'Clinton',
    1992: 'Clinton', 1988: 'Bush', 1984: 'Reagan', 1980: 'Reagan', 1976: 'Carter'
}

records = []
for row_idx in range(3, len(df)):
    state  = df.iloc[row_idx, 0]
    region = df.iloc[row_idx, 1]
    if pd.isna(state):
        continue
    for year, (d, r, ps) in year_col_map.items():
        records.append({
            'Year':              year,
            'State':             state,
            'Region':            region,
            'Democratic %':      df.iloc[row_idx, d],
            'Republican %':      df.iloc[row_idx, r],
            'P.S.':              df.iloc[row_idx, ps],
            'P.S. Score':        df.iloc[row_idx, ps_col_map[year]],
            'National Winner':   national_winners[year],
        })

result = pd.DataFrame(records).sort_values(['Year', 'State']).reset_index(drop=True)
result.to_csv('election_results_1976_2024.csv', index=False)
print(result)
```

```python
import hashlib

with open("election_results_1976_2024.csv", "rb") as f:
    sha256 = hashlib.sha256(f.read()).hexdigest()

with open("election_results_1976_2024.csv.sha", "w") as f:
    f.write(sha256)
```
