```python
# importing csv data 
import pandas as pd
df = pd.read_csv('election_results_1976_2024.csv')

df.head()
```

```python
df
```

```python
# Step 1: Strip % and convert
df['Year'] = df['Year'].astype(int)
df['Democratic %'] = pd.to_numeric(df['Democratic %'].str.replace('%', ''), errors='coerce')
df['Republican %'] = pd.to_numeric(df['Republican %'].str.replace('%', ''), errors='coerce')
df['P.S. Score'] = pd.to_numeric(df['P.S. Score'], errors='coerce')

# Step 2: Rename
df.rename(columns={
    'Democratic %': 'dem_pct',
    'Republican %': 'rep_pct',
    'P.S.': 'party_strength',
    'P.S. Score': 'ps_score',
    'National Winner': 'national_winner',
    'State Winner': 'state_winner'
}, inplace=True)
```
